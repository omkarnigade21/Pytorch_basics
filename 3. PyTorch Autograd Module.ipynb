```
# importing torch library
import torch
```

```
# initializing a 2d tensor
a = torch.tensor([[1,1], [1,1]], dtype=torch.float)
a
```

```
# performing operations on the tensor
b = a + 5
print(b)
```

```
# performing another operation
c = b.mean()
print(c)
```

## Automatic Differentiation

```
# initializing a 2d tensor and setting requires_grad as
# requires_grad = True used to calculate differentiation of any function wrt a

a = torch.tensor([[1,1], [1,1]], dtype= torch.float, requires_grad=True)
a
```

```
# performing operations on the tensor
b = a + 5
c = b.mean()
print(b)
print(c)
```

```
# back propagating
# gives rate of change of c with respect to backward variables.(Need to keep requires_grad = True for those backward wariables)
# Here backward variables for c are b and a as c = f(b), b = g(a)
c.backward()
```

```
# calculated gradients
print(a.grad)
```
